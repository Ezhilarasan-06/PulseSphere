```python

from pyspark.sql import SparkSession
spark = SparkSession.builder.getOrCreate()

csv_path = "/Workspace/Users/yalini.sathiya@sridsoftware.com/twitter_sentiment_dataset.csv"


df_raw = spark.read.option("header", "true") \
                   .option("inferSchema", "true") \
                   .csv(csv_path)

df_raw.printSchema()
display(df_raw.limit(10))
```

write a data

```python
from pyspark.sql.functions import current_timestamp, lit

df_bronze = df_raw.withColumn("_ingest_timestamp", current_timestamp()) \
                  .withColumn("_source", lit("twitter_sentiment_dataset.csv"))


df_bronze.write.format("delta") \
               .mode("overwrite") \
               .saveAsTable("workspace.default.twitter_raw")

print("Bronze table written in UC: workspace.default.twitter_raw")
```
