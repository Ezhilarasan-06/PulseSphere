```python
%sql
SHOW TABLES IN default;
```

Trending Topics View

```python
%sql
CREATE OR REPLACE VIEW vw_gold_trending_topics AS
SELECT
    hashtag AS topic,
    count AS total_mentions
FROM default.gold_trending_topics
ORDER BY total_mentions DESC;
```

Daily Sentiment Trend View

```python
%sql
CREATE OR REPLACE VIEW vw_gold_daily_sentiment AS
SELECT
    date,
    sentiment,
    sentiment_count AS total_posts
FROM default.gold_daily_sentiment
ORDER BY date;
```

```python
%sql
CREATE OR REPLACE VIEW vw_gold_sentiment_percentage AS
SELECT
    sentiment,
    percentage
FROM default.gold_sentiment_percentage
ORDER BY percentage DESC;
```

```python
%sql
CREATE OR REPLACE VIEW vw_gold_hashtag_trends AS
SELECT
    hashtag,
    total_count AS total_mentions
FROM default.gold_hashtag_trends
ORDER BY total_mentions DESC;
```

```python
%sql
CREATE OR REPLACE VIEW vw_gold_daily_hashtag AS
SELECT
    date,
    hashtag,
    hashtag_count
FROM default.gold_daily_hashtag
ORDER BY date, hashtag_count DESC;
```

```python
%sql
CREATE OR REPLACE VIEW vw_gold_top_users AS
SELECT
    user_id,
    username,
    total_likes,
    total_replies,
    engagement_score
FROM default.gold_top_users
ORDER BY engagement_score DESC;
```
